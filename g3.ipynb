```python
import pandas as pd
import numpy as np
from tabulate import tabulate

# Load the dataset
file_path = '/mnt/data/AI jobs by state.csv'
df_1 = pd.read_csv(r'C:\Users\madio\Documents\Group3 Project\group3-project-25.02.01\AI files\AI jobs by state.csv')

# Calculation of the mean of AI job postings
mean_value = df_1['Number of AI job postings in 2023'].mean()

# Difference from the mean of each state
df_1['Difference_from_Mean'] = abs(df_1['Number of AI job postings in 2023'] - mean_value)

# Sort by the closest states to the mean value
df_closest_to_mean = df_1.sort_values(by='Difference_from_Mean').head(10)

# Display the mean value and the table of states closest to the mean
print(f"Mean Number of AI Job Postings: {mean_value:.2f}")
print(tabulate(df_closest_to_mean, headers='keys', tablefmt='psql'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame({ 
    'State': ['DC', 'DE', 'MD', 'VA', 'MA', 'WA', 'CO', 'CA', 'NY', 'IL'],
    'Industry': ['Information', 'Scientific and Tech Services', 'Finance and Insurance', 'Manufacturing', 'Health Care and Social Assistance', 'Public Administration', 'Educational Services', 'Retail Trade', 'Construction', 'Construction'],
    'Year': [2021, 2022, 2023, 2023, 2021, 2022, 2023, 2023, 2021, 2022],
    'AI_Jobs': [5000, 3000, 4000, 2500, 4500, 3500, 3000, 2000, 1500, 1000],
    'Percentage_Growth': [10, 20, 15, 25, 30, 35, 40, 45, 50, 55],
    'Skills': ['Machine Learning', 'Artificial Intelligence', 'Natural language processing', 'Autonomous driving', 'Neural networks', 'Deep learning', 'Computer vision', 'Robotics', 'Data mining', 'Predictive analytics']
})

pivot_data = data.pivot(index='Industry', columns='State', values='AI_Jobs')
plt.figure(figsize=(10, 6))
heatmap = sns.heatmap(pivot_data, annot=True, fmt=".0f", cmap='YlGnBu')  

# Customize tick labels
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, ha="right", fontsize=12, fontweight='bold')
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0, fontsize=12, fontweight='bold')
labl = plt.ylabel('Industry', fontsize=14, fontweight='bold')
labl.set_color('red')
labl = plt.xlabel('State', fontsize=14, fontweight='bold')
labl.set_color('red')

# Use fmt=".0f" to format annotations as integers
plt.title('Heatmap of AI Jobs by Industry and State', fontsize=16, fontweight='bold', color='red')
plt.show()
```
