```python

import pandas as pd
from tabulate import tabulate

print("AI Jobs by percentage by state in 2023")
df_1 = pd.read_csv(r'C:\Users\madio\Documents\Group3 Project\group3-project-25.02.01\AI files\AI jobs by percentage by state.csv')
top_10_states = df_1.sort_values(by="Percentage of US states job postings in AI in 2023", ascending=False)
print(tabulate(top_10_states.head(10), headers='keys', tablefmt='psql'))

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a dictionary with the data
data = {
    "Geographic area": [
        "United States", "China", "India", "Germany", "United Kingdom",
        "Canada", "Australia", "France", "Japan", "South Korea"
    ],
    "AI job postings (% of all job postings)": [
        15.3, 12.8, 11.5, 9.7, 8.4, 7.5, 6.9, 5.8, 5.4, 4.2
    ]
}

# Create a DataFrame
df_1 = pd.DataFrame(data)

# Add a Year column
df_1['Year'] = 2023

# Sort the data by percentage for a clearer view
df_sorted = df_1.sort_values(by='AI job postings (% of all job postings)', ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(df_sorted['Geographic area'], df_sorted['AI job postings (% of all job postings)'], color='skyblue')

# Customize the chart
plt.title("AI Jobs by Percentage by Country in 2023", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Percentage of AI Job Postings", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the chart
plt.show()
```
